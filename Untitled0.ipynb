```python
from scipy.integrate import ode
from models import nmda, LR, LR_ICa, HH
%pylab inline
```

```python
def nmda_impulse(t, y, T=10, t_in=500, t_end=1500):
    rb = 0.005
    dy = nmda(y)
    if t_in <= t < t_end:
        dy[0] += -rb * T * y[0]
        dy[1] += rb * T * y[0] - rb * y[1] * T
        dy[2] += rb * y[1]
    return dy

def ica(O, V):
    Mg = 1.0
    gNMDA = 0.4 #+ 0.1*randn()
    VCa = 100.0
    B = 1.0 / (1 + exp(-0.062 * V)*Mg/3.57)
    ICa = gNMDA * B * O * (V - VCa)
    return ICa

def nmda_LR(t, y):
    dy = zeros_like(y)
    dy[:4] = nmda_impulse(t, y[:4])
    ICa = ica(y[3], y[4])
    dy[4:] = LR(y[4:], ICa)
    #dy[6] = 0.007*(-0.0275* ICa - 0.02*y[6]) 
    return dy
    
def nmda_HH(t, y):
    dy = zeros_like(y)
    dy[:4] = HH(y[:4])
    dy[4:] = nmda_impulse(t, y[4:])
    ICa = ica(y[7], y[0])
    dy[0] -= ICa
    return dy

solver = ['dopri5', 'dop853', 'vode']
```

#Rinzel and Lee
$C_m \frac{dv}{dt}=-I_{Ca}   −\left(g_K n^4 +\frac{g_{K, Ca} c}{K_d + c}\right) (V − V K ) − g_L (V − V_L )$

$\tau_n \frac{dn}{dt}=n_{\inf}(V)-n$

$\frac{dc}{dt} = f(-k_1 I_{Ca} -k_c c)$

```python
y0 = array([-50.0, 0.530, 0.7])
r = ode(lambda t, y: LR(y)).set_integrator(solver[0])
r.set_initial_value(y0, 0)
t1 = 20000.0
dt = 1.0
v = []
ca = []
time = []
while r.successful() and r.t < t1:
    r.integrate(r.t+dt)
    v.append(r.y[0])
    ca.append(r.y[2])
    time.append(r.t)
figure(1)
plot(time, v)
xlabel('timepo (s)')
ylabel('V (mV)')
grid(True)
figure(2)
plot(time, ca)
xlabel('timepo (s)')
ylabel('[Ca] (M)')
grid(True)

show()
```
